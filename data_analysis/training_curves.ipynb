```python
import glob

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8))
c_dict = {0:'r',1:'b',2:'g',3:'black',4:'cyan',5:'yellow',6:'orange',7:'purple',8:'lightblue',9:'orchid'}

file_names = glob.glob("training_graphs/ELO/*.csv")
df_files = pd.DataFrame(file_names, columns=['file_names'])
df_files['name'] = df_files['file_names'].str.replace(".csv","", regex=False).str.split("/").apply(lambda x:x[-1])
df_files = df_files[~df_files['name'].isin(['bs_1000','bs_default','POCA_half_ELO','bs_500000','bs_12000'])]
df_files['name'] = df_files['name'].apply(lambda x: 'POCA_' + x if 'bs_' in x else x)
i=0
for f, name in df_files[['file_names','name']].values:
    dat = pd.read_csv(f)
    plt.plot(dat['Step'], dat['Value'],c=c_dict[i],linestyle='-', label=name)
    i+=1

plt.legend()
plt.xlabel("Step", fontsize=18)
plt.ylabel("SelfPlay ELO", fontsize=18)
plt.title("Learning curve (SelfPlay ELO)", fontsize=20)
```
